## For the given ‘Iris’ dataset, create the Decision Tree classifier and visualize it graphically

## Importing the Required libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
iri = pd.read_csv('Iris.csv')
iri.head()
```

## Visual Comparision

```python
pd.plotting.parallel_coordinates(iri.iloc[:,1:],"Species", colormap=plt.get_cmap("Set2"))
plt.show
```

## Getting only the feature columns 

```python
iris = iri.drop(['Id','Species'],axis=1)
iris.head()
```

## Checking out our data

```python
iris.info()
```

```python
iri['Species'].value_counts()
```

## Getting the target array

## We have used LabelEncoding on 'Species' categorical feature

```python
from sklearn.preprocessing import LabelEncoder
enco = LabelEncoder()
target = enco.fit_transform(iri['Species'])
target
```

## Defining the decision tree algorithm

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier(random_state=42)
clf.fit(iris,target)

print("Decision Tree Classifier created")
```

## Visualizing the Tree

```python
fl = [x for x in iris.columns]
ll = ['Iris-setosa','Iris-versicolor','Iris-virginica']
```

```python
import matplotlib.pyplot as plt
from sklearn import tree
fig = plt.figure(figsize=(20,25))
_ = tree.plot_tree(clf, 
                   feature_names= fl,  
                   class_names= ll,
                   filled=True)
```
